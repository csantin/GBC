```python
import pandas as pd

# Get train and test data
df_train = pd.read_json('train.json')
df_test = pd.read_json('test.json')
```

```python
#Transform ingredients from list of list of strings to list of strings
df_train['ingstring'] = df_train['ingredients'].map(lambda x: ','.join(x))
df_train.ingstring
df_test['ingstring'] = df_test['ingredients'].map(lambda x: ','.join(x))
df_test.ingstring
```

```python
# nltk import and test
import nltk
nltk.download('punkt')
nltk.word_tokenize("milk,salt,(4)dried red chile peppers,ground red")
```

```python
# Set up stop words to be removed
remove = ['leaves', 'powder', 'artificial', 'fried', 'black', 'barbequed', 'cold', 'flakes', 'grated', 'dry', 'pure', 'fingerling', 'chinese', 'white', 'granulated', 'coarse', 'green', 'red', 'refried', 'fat-free', 'firm', 'firmly', 'packed', 'rolled', 'boiling', 'yellow', 'freshly', 'fillet', 'skimmed', 'crusty', 'seeds', 'finely', 'accompaniment', 'accent', 'smoked', 'whipped', 'sliced', 'half', '&', 'American', 'stewed', 'sprigs', 'meat', 'parboiled', 'light', 'canned', 'heavy', 'low-fat', 'reduced', 'table', 'fresh', 'all-purpose', 'minced', 'fat', 'free', 'less', 'roasted', 'peeled', 'slivered', 'large', 'unsalted', 'dried', 'ground', 'extra-virgin', 'ripe', 'fat', 'sodium', 'chopped', 'whole', 'cracked', 'flat', 'kosher', 'sharp', 'shredded']
# Remove words
def ingClean(ingString):
    ingredients = nltk.word_tokenize(ingString)
    newIng = ''
    for word in ingredients:
        if word not in remove and word.isalpha() and ' ' not in word:
            if newIng == '':
                newIng += word
            else:
                newIng += ',' + word
    return newIng

# Create new list of strings for ingredients lacking the stop words
df_train['inglist'] = df_train['ingstring'].map(lambda x: ingClean(x))
df_test['inglist'] = df_test['ingstring'].map(lambda x: ingClean(x))
df_train['inglist']

# Remove words (old)
# ilist = df_train.inglist

# counter = 0
# for words in ilist:
#     wordslist = words.split(',')
#     newwords = ''
#     for word in wordslist:
#         wordlist = word.split(' ')
#         newword = ''
#         for each in wordlist:
#             if each not in remove:
#                 newword += ' ' + each
# #                 words = words.replace(word, newword)
#         newword = newword.strip()
#         if newwords == '':
#              newwords += newword
#         else:
#             newwords += ',' + newword
#     ilist[counter] = newwords
#     counter += 1
# ilist
# df_train['inglist'] = ilist
```

```python
# Create vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(df_train.inglist)
# vectorizer.get_feature_names()

# Use encoder for cuisines

# Confirm vectorizer
vectorizer.get_feature_names()
len(vectorizer.get_feature_names())

#Finding Onions and getting Tfid value (test)
# onions = vectorizer.transform(['onions'])
# onions.toarray().argmax()
# onions[0,1775]
```

```python
# Using 5.5 Naive Bayes from Semester 1 and 3.6 from Semester 2

# Vectorizing train and test data, converting to standard array
X_train = vectorizer.transform(df_train.inglist).toarray()
X_test = vectorizer.transform(df_test.inglist).toarray()
```

```python
# Vectorize cuisines, converting to standard array
df_train.cuisine
from sklearn.feature_extraction.text import CountVectorizer

cuisine_vectorizer = CountVectorizer()
cuisine_vectorizer.fit(df_train.cuisine)
y_train = cuisine_vectorizer.transform(df_train.cuisine).toarray()
```

```python
# Build network
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(20, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
y_train.shape
```

```python
# Create validation data
X_val = X_train[:1000]
X_partial = X_train[1000:]
y_val = y_train[:1000]
y_partial = y_train[1000:]

# Train data
history = model.fit(X_partial,
                    y_partial,
                    epochs=20,
                    batch_size=512,
                    validation_data=(X_val, y_val))
```

```python
# Create prediction for test data
y_test = model.predict(X_test)
```

```python
# Checking test data
import numpy as np
y_test[0]
np.sum(y_test[0])
np.argmax(y_test[0])
```

```python
# Converting vectorized result array into boolean for inverse transform
y_result = []
for i in y_test:
    arr = i
    max = np.argmax(i)
    choice = list(map(lambda x: x == i[max], arr))
    y_result.append(cuisine_vectorizer.inverse_transform(choice)[0][0])
```
